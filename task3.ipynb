```python
from sklearn.datasets import load_digits
import pandas as pd
from matplotlib import pyplot as plt
```

```python
digits=load_digits()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(x_train,y_train)
```

```python
x = range(len(digits.data[5]))  # Index positions for the features
y = digits.data[5]  # Features of the 10th sample
plt.scatter(x, y, marker="*")
plt.plot(x,color="red", label="Line Through Points")
plt.xlabel("Feature Index")
plt.ylabel("Feature Value")
plt.title("Feature Values for the 10th Sample")
plt.show()
print("model scores:",l.score(x_test,y_test))
```
